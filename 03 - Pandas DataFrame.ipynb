data Frame veri yapısı birden fazla veri yapısını aynı anda içeren toplolarda diyebiliriz.

```python
import pandas as pd
```

```python
# Birbirinden farklı türdeki verileri bu şekilde "dateframe" veri yapısı ile görselleştirebilirim.
 
veri = {
    "isim": ["Ali","Ahmet","Efe","Kemal","Ayşe","Aslı"],
    "spor": ["Futbol","Yüzme","Basketbol","Güreş","Bale","Voleybol"],
    "cinsiyet": ["E","E","E","E","K","K"],
    "puan": [75,70,80,90,80,85]
}
df = pd.DataFrame(veri)
df
```

Şimdi bu veri tipinde verileri görselleştirmede kullanacağımız methotlara bakalım.

```python
# Eğer verinin ilk beş verisini gertirmek istersem bu şekilde yapmam gerekeçektir.

df.head()
```

```python
# Eğer son beş satırı ekrana yazdırmak istersem ise bu şekilde yapmam gerekeçektir.

df.tail()
```

```python
# Eğer sütünların sırasını değiştirmek istersem ise yapmam gerek şu olmalıdırç

df = pd.DataFrame(veri,columns=["isim","cinsiyet","spor","puan"])
df
```

```python
# Eğer yeni sütün ekleyip bu sütünün veri girişini yapmazsam bu sütündaki veriler eksik veri olarak tanımlanmaktadır.

df = pd.DataFrame(veri,columns=["isim","cinsiyet","spor","puan","yaş"])
df
```

```python
# Eğer indexlere etiket atamak istersem yapmam gerek şu olmalıdır.

df = pd.DataFrame(veri,columns=["isim","cinsiyet","spor","puan","yaş"],
                 index=["bir","iki","uc","dort","bes","altı",])
df
```

```python
# Eğer belirli bir sütünü almal istersek yapmamız gereken şu olmalırdır.

df["spor"]
```

```python
# Birden fazla belirli sütünü almak istersem yapmam gereken şu olmalıdır.

sutunlarım = ["isim","puan"]
df[sutunlarım]
```

```python
# İstediğim sütünlara başka bir yoldanda ulaşabilirim.

df.spor
```

```python
# Belirli bir satırı çağırmak istediğimizde şu şekil çağırmak zorundayım.

df.loc[["bir"]]
```

```python
# Belirli birden fazla satır çağırmak istesem.

df.loc[["bir","uc"]]
```

```python
# Değeri olmayan bir sütüna değer ataması yapmak istersem ise bu değerleri nasıl atamasını yapmam gerektiğini şu şekil yapmak mecburiyetindeyim.
# Benim örneğimde bu süün yaş sütününa denk gelmektadir.

df["yaş"]=18
df
```

```python
# Eğer bir dizi şeklinde atama yapmam gerekir ise şu şekilde olmalıdır.

deger = [18,17,18,19,19,17]
df["yaş"] = deger
df
```

```python
# dersten 70 ve üzeri alanları "üstün başarı" adında bir ünvan vermek istiyorum.

df["Üstün başarı"] = df.puan>70
df
```

```python
# Eğer bir satır silmek istersen ise yapmam gereken şu olmalıdır.

del df["Üstün başarı"]
df
```

```python
# İç içe sözlük yapısı oluşturmak istediğimde ise yapmam gereken şu olmalıdır.
# Üst alandaki sözlüğü burası içinde kullanabilirim.
# Pandas modülü dışdaki sözlüğü sütün olarak içteki sözlüğü satır olarak almaktadır.

notlar = {
    "matematik": {"Ali":85,"Ahmet":80,"Efe":75,"Kemal":90},
    "fizik": {"Ali":80,"Ahmet":75,"Efe":70,"Kemal":90}
}
puan = pd.DataFrame(notlar)
puan
```

```python
# DataFrame yapısında satırlar ve sütünlar yer değiştirile bilir bunu ise şu şekilde yapmamız gerekmektedir.

puan.T
```

```python
# DataFrame yapısında satır ve sütünlara isim verebilirim.

puan.index.name="isim"
puan.columns.name="ders"
puan
```

```python
# Eğer verileri iki boyutlu olarak görmek istersem yapmam gereken şu olmalıdır.

puan.values
```

```python
# Eğer bir veriyi değiştirilemez hale getirmek istiyorsam gereken şu olmalıdır.

indeks=puan.index
```
